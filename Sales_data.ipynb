Sales Data Analysis

### Gender

```python
# plotting a bar chart for Gender and it's count
df.head(2)
```

```python
g = sns.countplot(x = 'Gender',data = df, hue="Gender")

for bars in g.containers:
    g.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

sd = df.groupby(["Gender"], as_index = False)["Amount"].sum().sort_values(by="Amount", ascending = False)

sns.barplot(data = sd, x = "Gender", y = "Amount",hue = "Gender")
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
df.head(2)
```

```python
az = sns.countplot(data = df, x = "Age Group",hue = "Gender")

for bars in az.containers:
    az.bar_label(bars)
```

```python
# Total Amount vs Age Group
df.head(2)
```

```python
sales_age = df.groupby(["Age Group"], as_index = False)["Amount"].sum().sort_values("Amount",ascending =  False)

sns.barplot(data = sales_age, x = "Age Group", y = "Amount", hue = "Age Group")
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### State

```python
# total number of orders from top 10 states
df.head(2)
```

```python
aw = df.groupby(["State"], as_index = False)["Orders"].sum().sort_values( "Orders", ascending = False).head(10)

sns.set({'figure.figsize':(15,5)})
sns.barplot(data = aw, x = "State", y = "Orders" , hue = "State")
```

```python
# total amount/sales from top 10 states
df.head(2)
```

```python
state_sales = df.groupby(["State"],  as_index = False)["Amount"].sum().sort_values("Amount",ascending = False).head(10)

sns.barplot(data = state_sales, x = "State", y = "Amount", hue = "State")
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Marital Status

```python
sns.set({"figure.figsize" : (6,5)})
ar = sns.countplot(data = df, x = "Marital_Status",hue = "Marital_Status")

```

```python
ae = df.groupby(["Marital_Status", "Gender"], as_index = False)["Orders"].sum().sort_values("Marital_Status",ascending = False).head(10)

sns.set({'figure.figsize': (6,5)}) 
sns.barplot(data = ae, x = "Marital_Status", y = "Orders", hue = "Gender")
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### Occupation

```python
sns.set({"figure.figsize": (20,5)}) 
sns.countplot(data = df, x = "Occupation", hue = "Occupation")
```

```python
ocup_sales = df.groupby(["Occupation"], as_index = False)["Amount"].sum().sort_values("Amount", ascending = False)

sns.barplot(data = ocup_sales, x = "Occupation", y = "Amount", hue = "Occupation")
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
sns.set({"figure.figsize": (22,5)}) 
aq = sns.countplot(data = df, x = "Product_Category", hue = "Product_Category")

for bars in aq.containers:
    aq.bar_label(bars)
```

```python
pd_sales = df.groupby(['Product_Category'],as_index = False)["Amount"].sum().sort_values("Amount", ascending = False).head(10)

sns.set({"figure.figsize" : (17,5)})
sns.barplot(data = pd_sales, x = "Product_Category", y = "Amount",hue = "Product_Category")
```

```python
pd_orders = df.groupby(['Product_Category'], as_index = False)['Orders'].sum().sort_values(["Orders"], ascending = False).head(10)

sns.barplot(data = pd_orders, x = "Product_Category", y = "Orders", hue = "Product_Category")
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
pi_orders = df.groupby(["Product_ID"], as_index = False)["Orders"].sum().sort_values("Orders", ascending = False).head(10)

sns.barplot( data = pi_orders, x = "Product_ID", y = "Orders", hue = "Product_ID")
```
